```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/Users/karlniwong/Documents/Python/strava_activities.csv')
df.drop(['Unnamed: 0'], 1)
df.head()
```

```python
def remove_time(text):
    return text[0:10]
date_format = "%Y-%m-%d"

df['start_date_local'] = df['start_date_local'].apply(remove_time) # convert to date format
df['start_date_local'] = pd.to_datetime(df['start_date_local'], format = date_format)
print(len(df))
df.head()
```

```python
year = '2023-01-01'

df_year = df[~(df['start_date_local'] < year)]
len(df_year)
```

```python
# sum everything up
df_run = df_year[df_year['type'] == 'Run']
len(df_run)
```

```python
total_running_dist = np.round(np.sum(df_run['distance'])/1000, 1) # in kilometers
total_running_time = int(np.ceil(np.sum(df_run['moving_time'])/3600))
total_running_elevation = int(np.ceil(np.sum(df_run['total_elevation_gain'])))
```
